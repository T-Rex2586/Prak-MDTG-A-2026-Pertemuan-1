# 1 Set atau himpunan
*   kumpulan element dengan nilai berbeda
*   unordered, tidak bisa dicari dengan indeks
*   kompleksitas waktu, O(1)




# 1.1 Membuat Set di python
Set di python bisa dibuat dengan membuat variabel dengan kumpulan nilai yang diapit dengan lambang '{}' atau buat manual sendiri fungsinya dari awal

```python
matkul = {'matdis','basdat','matlan','pnacasila'}
print(matkul)
```

# 1.2 Menambah element set
  **add**, menambah satu element kedalam set

```python
matkul.add('100')
print(matkul)

```

**update**, menambah beberapa element kedalam set

```python
matkul.update(['101','102'])
print(matkul)
```

# 1.3 Menghapus element set
  **discard**, menghapus element dalam set tanpa pesan error jika element tidak ada

```python
matkul.discard('100')
print(matkul)

matkul.discard('100')
print(matkul)
```

**remove**, menghapus element dalam set dengan pesan error jika element tidak ada

```python
matkul.remove('101')
print(matkul)

matkul.remove('101')
print(matkul)
```

# 2.1 Union

```python
set_1 = {5,4,3,2,1}
set_2 = {1,3,5,6}
set_3 = {10,11,12,13}
set_4 = {1,2,3}
print(set_1 | set_2)
#atau
print(set_1.union(set_2))
```

# 2.2 Intersection

```python
print(set_1 & set_2)
#atau
print(set_1.intersection(set_2))
```

# 2.3 Difference

```python
print(set_1 - set_2)
print(set_2 - set_1)
#atau
print(set_1.difference(set_2))
print(set_2.difference(set_1))
```

# 2.4 symmetric difference
  nilai dari dua atau lebih set yang ada diluar irisan

```python
print(set_1 ^ set_2)
#atau
print(set_1.symmetric_difference(set_2))
```

# 2.5 disjointset
  ngasih tau set yang dibandingkan apakah seluruh elementnya berbeda atau tidak. True berarti semuanya berbeda, False berarti ada yang sama

```python
print(set_1.isdisjoint (set_2))
print(set_1.isdisjoint (set_3))
```

# 2.6 Subset dan superset

Subset → semua elemen set kecil ada di set besar <br>
Superset → set besar mengandung semua elemen set kecil

```python
print(set_1.issubset(set_4))
print(set_4.issubset(set_1))
```

cara singkat menggunakan operasi <=

```python
print(set_1 <= set_4)
print(set_4 <= set_1)
```

```python
print(set_1.issuperset(set_4))
print(set_4.issuperset(set_1))
```

cara singkat menggunakan operasi >=

```python
print(set_1 >= set_4)
print(set_4 >= set_1)
```

#3. Membership
Membership menyatakan ada atau tidaknya sebuah elemen didalam suatu set

Dalam python, Membership dapat di cek menggunakan in atau not in

```python
matkul = {'matdis','basdat','matlan','pnacasila'}
```

```python
allin = "Allin" in matkul
basdat = "basdat" in matkul

print(f"Allin in matkul   : {allin}")
print(f"Basdat in matkul  : {basdat}")
```

```python
allin = "Allin" not in matkul
basdat = "basdat" not in  matkul

print(f"Allin not in matkul   : {allin}")
print(f"Basdat not in matkul  : {basdat}")
```

#4.Cardinality
Cardinality atau Kardinalitas menyatakan jumlah elemen dalam sebuah set.

Pada python, Cardinality dapat dicari menggunakan fungsi len()

```python
set_angka = {1,2,3,4,5,6,7,8,9}
```

```python
print(f"Jumlah Elemen Set Angka = {len(set_angka)}")
```

#5.Komplemen
Komplemen merupakan seluruh anggota semesta yang tidak termasuk kedalam set yang dipilih.

Cara mencari komplemen dapat menggunakan Operasi -

```python
semesta = {1,2,3,4,5,6,7,8,9}
set_angka = {4,5,6}

set_angka_komplemen = semesta - set_angka

print(set_angka_komplemen)
```
