## 1. Visualisasi Relasi

```python
import matplotlib.pyplot as plt
import numpy as np

def visualisasi_relasi (Set1,Set2,Relasi={},Judul="",NamaSet1="",NamaSet2="",WarnaSet1="skyblue",WarnaSet2="lightgreen") :
    fig, ax = plt.subplots(figsize=(8, 6))
    pos_x_Set1 = np.zeros(len(Set1))
    pos_x_Set2 = np.ones(len(Set2))
    pos_y_Set1 = np.linspace(1,5,len(Set1))
    pos_y_Set2 = np.linspace(1,5,len(Set2))
    ax.scatter(pos_x_Set1, pos_y_Set1, s=500, c=WarnaSet1, edgecolors='black', zorder=2)
    ax.scatter(pos_x_Set2, pos_y_Set2, s=500, c=WarnaSet2, edgecolors='black', zorder=2)
    koor_Set1 = {}
    koor_Set2 = {}
    
    indeks = 0
    for i in Set1:
        ax.text(-0.1, pos_y_Set1[indeks], str(i), fontsize=12, ha='right', va='center', fontweight='bold')
        koor_Set1[str(i)] = (0,pos_y_Set1[indeks])
        indeks += 1
        
    indeks = 0
    for i in Set2:
        ax.text(1.1, pos_y_Set2[indeks], str(i), fontsize=12, ha='left', va='center', fontweight='bold')
        koor_Set2[str(i)] = (1,pos_y_Set2[indeks])
        indeks += 1
        
    for i in Relasi:
        for j in Relasi[i]:
            if i in Set1 and j in Set2 :
                ax.annotate("", 
                            xy=koor_Set2[str(j)],    
                            xytext=koor_Set1[str(i)], 
                            arrowprops=dict(arrowstyle="-", connectionstyle="arc3", color='black', lw=1.5),
                            zorder=1)
                
    ax.set_title(Judul, fontsize=16, pad=20)
    ax.set_xticks([0, 1])
    ax.set_xticklabels([NamaSet1, NamaSet2], fontsize=14, fontweight='bold')
    ax.set_yticks([])
    ax.set_xlim(-0.5, 1.5)
    ax.set_ylim(0, 6)
    
    plt.tight_layout()
    plt.show()
    
```

```python
Asprak = {"Elias","Rexy","Alvin","Hana"}
Prodi = {"Sains Data", "Informatika"}
Relasi = {
    "Elias" : ["Sains Data"],
    "Rexy" : ["Sains Data"],
    "Alvin" : ["Sains Data"],
    "Hana" : ["Informatika"]
}

visualisasi_relasi(Set1=Asprak , Set2=Prodi , Relasi=Relasi , NamaSet1="Asprak" , NamaSet2="Prodi", Judul="Prodinya Para Asprak Matdis")
```

## 2. Rekursi

#### 2.1 Pengetian Rekursi

Rekursi adalah sebuah teknik di mana suatu fungsi memanggil dirinya sendiri secara berulang-ulang untuk menyelesaikan sebuah masalah.

Kamu bisa membayangkan rekursi seperti boneka Matryoshka (boneka Rusia). Untuk mencapai boneka terkecil di bagian dalam, kamu harus membuka boneka yang lebih besar satu per satu. Setiap kali kamu membuka boneka, kamu melakukan tindakan yang sama, hanya saja pada boneka yang ukurannya lebih kecil.

Agar sebuah fungsi rekursif tidak berjalan tanpa henti (infinite loop) dan menyebabkan stack overflow (memori penuh), fungsi tersebut wajib memiliki dua komponen utama:
1. Base Case (Kondisi Dasar/Berhenti): Ini adalah kondisi di mana fungsi berhenti memanggil dirinya sendiri dan langsung mengembalikan sebuah nilai. (Ibarat boneka Matryoshka terkecil yang tidak bisa dibuka lagi).
2. Recursive Case (Langkah Rekursif): Ini adalah bagian di mana fungsi memanggil dirinya sendiri, tetapi dengan input yang nilainya sudah dimodifikasi mendekati base case.

Pengaplikasian rekursi: fibonacci, membuat binary tree, rekursi relations, dan lainnya.

#### 2.2 Contoh Pengaplikasian Rekursi

Deretan Fibonacci adalah barisan angka yang bentuk bilangannya adalah 0, 1, 1, 2, 3, 5, 8, 13, 21, dan seterusnya. Pola utamanya adalah setiap elemen ke-$n$ merupakan jumlah dari dua elemen sebelumnya. Bentuk umum deretan Fibonacci dapat dituliskan secara rekursif sebagai berikut:
- Basis: $F_0 = 0$ dan $F_1 = 1$
- Rekurens: $F_n = F_{n-1} + F_{n-2}$ untuk $n \ge 2$

```python
def fibonacci(n):
    # 1. Basis
    if n == 0:
        return 0
    elif n == 1:
        return 1
    # 2. Rekurens
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

# Menguji fungsi untuk mencari Fibonacci ke-10
n_target = 10
hasil = fibonacci(n_target)
print(f"Angka Fibonacci ke-{n_target} adalah {hasil}")
```

#### 2.3 Studi Kasus

- Bayangkan kamu memiliki 5 buah buku berbeda dan ingin menyusunnya berjejer di rak. Berapa banyak kombinasi susunan buku yang bisa kamu buat? <br>
hint: dapat diselesaikan dengan faktorial

```python
def hitung_faktorial(n):
    # 1. Basis (Kondisi Berhenti)
    if n == 0:
        return 1
    # 2. Rekurens (Langkah Rekursif)
    else:
        return n * hitung_faktorial(n - 1)
    

jumlah_buku = 8
kombinasi = hitung_faktorial(jumlah_buku)
print(f"Ada {kombinasi} cara susunan berbeda untuk {jumlah_buku} buku.")
```

## 3. Recurrence Relations

### 3.1 Pengertian

Relasi rekurens adalah sebuah persamaan yang mendefinisikan sebuah barisan secara rekursif, di mana setiap suku dalam barisan tersebut dinyatakan sebagai fungsi dari suku-suku sebelumnya. Dalam matematika diskrit, relasi rekurens sangat penting untuk memodelkan masalah yang memiliki struktur serupa di setiap tahapannya, seperti pertumbuhan populasi atau kompleksitas algoritma.

### 3.2 Penyelesaian Recurrence Relations

Penyelesaian rekurens terbentuk sesuai dengan jenisnya:
- Homogeneous recurrence
- Non-Homogenous recurrence

Homogeneous recurrence adalah relasi rekurens linier di mana setiap sukunya merupakan kombinasi linier dari suku-suku sebelumnya dan disamakan dengan nol (tidak memiliki fungsi tambahan atau konstanta yang berdiri sendiri).
Bentuk umumnya adalah:

$$a_n + c_1a_{n-1} + c_2a_{n-2} + \dots + c_ka_{n-k} = 0$$

Sebagai pelengkap, Non-Homogeneous recurrence adalah relasi rekurens yang memiliki tambahan sebuah fungsi $f(n)$ yang tidak bergantung pada elemen barisan tersebut.
Bentuk umumnya adalah:

$$a_n + c_1a_{n-1} + c_2a_{n-2} + \dots + c_ka_{n-k} = f(n)$$

#### 3.2.1 Penyelesaian Homogenous Recurrence

Contoh soal:
$$a_n = 6a_{n-1} - 9a_{n-2}$$
with initial condition $$a_1=3 \quad \text{dan} \quad a_2=27$$

##### Metode Akar Karakteristik (Characteristic Roots)

Sebenarnya metodenya ada banyak seperti generating function, telescoping, dll. Tetapi kita hanya belajar menggunakan 1 metode ini. Konsep dari metode ini adalah mengasumsikan solusi berbentuk $a_n = r^n$, kemudian mencari nilai r yang memenuhi persamaan.

Penyelesaian:
1. Bentuk persamaan karakteristik: $r^2 - 6r + 9$
2. Cari akarnya: Hanya terdapat 1 akar yaitu $r_0 = 3$
3. Maka general solution adalah $a_n = c_1{3^n} + c_1{n}{3^n}$ 
4. Cari value $c$ dengan memasukkan persamaan awal <br>
   Untuk $n=1$, maka $c_1{3^1} + c_1{(1)}{3^1} = 2$ <br>
   Untuk $n=2$, maka $c_1{3^2} + c_1{(2)}{3^2} = 27$
5. Maka hasilnya akan ketemu $c_1=-1$ dan $c_2=2$
6. Unique solution: $a_n = 3^n(2n-1)$

*Ingat jika terdapat akar yang sama, maka harus ada penambahan $n$ di belakang

```python
import sympy as sp

def temukan_unique_solution():
    # 1. Definisikan 'n' sebagai simbol matematika (integer) dan 'y' sebagai fungsi
    n = sp.symbols('n', integer=True)
    y = sp.Function('y')
    
    # 2. Tuliskan persamaan relasi rekurensinya
    # a_n - 6a_{n-1} + 9a_{n-2} = 0
    persamaan = sp.Eq(y(n) - 6 * y(n-1) + 9 * y(n-2), 0)
    
    # 3. Masukkan kondisi awal (Initial Condition)
    # Diketahui a_0 = 3, a_2 = 27
    kondisi_awal = {y(1): 3, y(2) : 27}
    
    # 4. Minta Python mencari "Unique Solution" 
    rumus_tertutup = sp.rsolve(persamaan, y(n), kondisi_awal)
    
    print(f"Persamaan Rekurensi : a_n = 6a_{n-1} - 9a_{n-2}")
    print(f"Rumus Unique Solution: y(n) = {rumus_tertutup}")
    
    # 5. BONUS: Mengubah rumus matematika tersebut kembali menjadi fungsi Python biasa
    # Ini disebut 'lambdify', mengubah ekspresi simbolik jadi fungsi yang bisa dieksekusi
    fungsi_baru = sp.lambdify(n, rumus_tertutup, 'math')
    
    return fungsi_baru

# Dapatkan fungsinya
unique_solution = temukan_unique_solution()
```

```python
# With general recurrence relation:
def fungsi_rekursi(n):
    if n == 1:
        return 3
    elif n == 2:
        return 27
    else:
        return 6 * fungsi_rekursi(n - 1) - 9 * fungsi_rekursi(n - 2)
```

```python
# Test perbandingan hasil
n_test = 1
for i in range(n_test, 6):
    print(f"a{i} = {int(fungsi_rekursi(i))}") # Kompleksitas O(2^n)
    print(f"a{i} = {int(unique_solution(i))}") # Kompleksitas O(1)

```

#### 3.2.2 Penyelesaian Non-Homogenous Recurrence

Contoh Soal 1: $$a_n = 2a_{n-1}- 1$$ With initial condition $a_0 = 1$

##### Metode Generating Function

Metode Generating Function (Fungsi Pembangkit) mengubah masalah barisan diskrit menjadi aljabar fungsi kontinu. Berikut adalah langkah-langkah detail penyelesaiannya:

**1. Definisikan Fungsi Pembangkitnya** <br>
Misalkan $A(x)$ adalah fungsi pembangkit dari barisan $a_n$, maka:
$$A(x) = \sum_{n=0}^{\infty} a_n x^n$$

**2. Kalikan Persamaan dengan $x^n$ dan Jumlahkan**<br>
Mulai dari persamaan awal:
$$a_n = 2a_{n-1} - 1$$

Karena relasi ini berlaku untuk $n \ge 1$, kita kalikan kedua ruas dengan $x^n$ lalu jumlahkan dari $n=1$ sampai tak hingga ($\infty$):
$$\sum_{n=1}^{\infty} a_n x^n = \sum_{n=1}^{\infty} 2a_{n-1} x^n - \sum_{n=1}^{\infty} 1 \cdot x^n$$
$$\sum_{n=1}^{\infty} a_n x^n = 2 \sum_{n=1}^{\infty} a_{n-1} x^n - \sum_{n=1}^{\infty} x^n$$

**3. Ubah Deret Menjadi Bentuk $A(x)$**<br>
Sekarang kita terjemahkan masing-masing bagian ke dalam fungsi $A(x)$:

* **Ruas Kiri:** Deret dimulai dari $n=1$, bukan $n=0$. Jadi ini adalah $A(x)$ dikurangi suku pertamanya ($a_0$). Karena diketahui di soal $a_0 = 1$, maka:
    $$\sum_{n=1}^{\infty} a_n x^n = A(x) - a_0 = A(x) - 1$$

* **Suku Pertama Ruas Kanan:** Kita bisa keluarkan satu $x$ agar indeksnya kembali ke $n=0$.
    $$2 \sum_{n=1}^{\infty} a_{n-1} x^n = 2x \sum_{n=1}^{\infty} a_{n-1} x^{n-1} = 2x \sum_{n=0}^{\infty} a_n x^n = 2x A(x)$$

* **Suku Kedua Ruas Kanan:** Deret $\sum_{n=1}^{\infty} x^n$ adalah deret geometri tak hingga yang rasionya $x$ dan suku pertamanya $x$. Rumusnya adalah $\frac{\text{suku pertama}}{1 - \text{rasio}}$.
    $$\sum_{n=1}^{\infty} x^n = \frac{x}{1-x}$$

**4. Susun Persamaan Aljabar dan Cari $A(x)$**<br>
Substitusikan kembali hasil langkah 3 ke dalam persamaan pada langkah 2:
$$A(x) - 1 = 2x A(x) - \frac{x}{1-x}$$

Pindahkan semua suku yang mengandung $A(x)$ ke satu sisi:
$$A(x) - 2x A(x) = 1 - \frac{x}{1-x}$$
$$A(x)(1 - 2x) = \frac{1-x}{1-x} - \frac{x}{1-x}$$
$$A(x)(1 - 2x) = \frac{1 - 2x}{1-x}$$

Bagi kedua ruas dengan $(1 - 2x)$:
$$A(x) = \frac{1}{1-x}$$

**5. Ubah Kembali ke Bentuk Barisan (Ekspansi Deret)**<br>
Sekarang kita telah mendapatkan bentuk akhir Fungsi Pembangkitnya, yaitu $A(x) = \frac{1}{1-x}$.
Kita tahu bahwa bentuk dasar deret geometri (deret Maclaurin) dari $\frac{1}{1-x}$ adalah:
$$\frac{1}{1-x} = 1 + x + x^2 + x^3 + \dots = \sum_{n=0}^{\infty} x^n$$

Karena $A(x) = \sum_{n=0}^{\infty} a_n x^n$ dan hasilnya sama persis dengan $\sum_{n=0}^{\infty} 1 \cdot x^n$, maka koefisien $a_n$ untuk setiap $x^n$ bernilai $1$.

**Solusi Akhir:**
Formula eksplisit (tertutup) untuk relasi rekurensi tersebut adalah:
$$a_n = 1$$

```python
import sympy as sp

def temukan_unique_solution():
    # 1. Definisikan 'n' sebagai simbol matematika (integer) dan 'y' sebagai fungsi
    n = sp.symbols('n', integer=True)
    y = sp.Function('y')
    
    # 2. Tuliskan persamaan relasi rekurensinya
    # Persamaan awal: a_n = 2a_{n-1} - 1
    persamaan = sp.Eq(y(n) - 2 * y(n - 1), -1)
    
    # 3. Masukkan kondisi awal (Initial Condition)
    # Diketahui a_0 = 1
    kondisi_awal = {y(0): 1}
    
    # 4. Minta Python mencari "Unique Solution" 
    rumus_tertutup = sp.rsolve(persamaan, y(n), kondisi_awal)
    
    print(f"Persamaan Rekurensi : y(n) = 2*y(n-1) - 1")
    print(f"Rumus Unique Solution: y(n) = {rumus_tertutup}")
    
    # 5. BONUS: Mengubah rumus matematika tersebut kembali menjadi fungsi Python biasa
    # Ini disebut 'lambdify', mengubah ekspresi simbolik jadi fungsi yang bisa dieksekusi
    fungsi_baru = sp.lambdify(n, rumus_tertutup, 'math')
    
    return fungsi_baru

# Dapatkan fungsinya
hitung_unique_solution = temukan_unique_solution()
```

```python
# 1. Iterative Style
def hitung_iterasi(n):
    a = 1
    an = 1 # Inisialisasi awal untuk menangani kasus n = 0
    
    for i in range(1, n + 1):
        an = 2 * a - 1
        a = an
        
    return an

# 2. Recursive Style
def hitung_rekursi(n):
    if n == 0:
        return 1
    else:
        return 2 * hitung_rekursi(n - 1) - 1


# Blok Pengujian (Test Code)
n_test = 5
    
print(f"Menguji untuk n = {n_test}:")
print(f"Hasil Iterasi        : {hitung_iterasi(n_test)}") # Kompleksitas O(n)
print(f"Hasil Rekursi        : {hitung_rekursi(n_test)}") # Kompleksitas O(2^n)
print(f"Hasil Unique Solution: {hitung_unique_solution(n_test)}") # Kompleksitas O(1)
```

### 3.3 Pengaplikasian Recurrence Relation

#### Pengaplikasian Generating Function untuk menghitung pi

```python
import math

def pi_1(N):
    jumlah = 0
    for n in range(1, N + 1):
        jumlah += 1 / (n ** 2)
    return math.sqrt(6 * jumlah)

def pi_2(N):
    jumlah = 0
    for n in range(1, N + 1):
        jumlah += 1 / ((2 * n) ** 2)
    return math.sqrt(24 * jumlah)

def pi_3(N):
    jumlah = 0
    for n in range(1, N + 1):
        jumlah += 1 / ((2 * n - 1) ** 2)
    return math.sqrt(8 * jumlah)

def pi_4(N):
    jumlah = 0
    for n in range(1, N + 1):
        jumlah += ((-1) ** (n + 1)) / (n ** 2)
    return math.sqrt(12 * jumlah)

if __name__ == "__main__":
    for N in range(1, 551, 50):
        print(f"{N}\t{pi_1(N):.6f}\t{pi_2(N):.6f}\t{pi_3(N):.6f}\t{pi_4(N):.6f}")
```
